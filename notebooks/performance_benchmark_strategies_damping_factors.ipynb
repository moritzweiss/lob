```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import os
base = os.path.dirname('/u/weim/lob/notebooks/performance_benchmark_strategies_damping_factors.ipynb')
parent_dir = os.path.dirname(base)
```

```python
# %%capture 
# Define the time array
t = np.linspace(0, 10, 100)

# Define the exponential damping function
plt.figure(figsize=(10, 6))
plt.title('Exponential Damping Function')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.xticks(np.arange(min(t), max(t)+1, 1.0))  # Set x ticks at every integer

damping_factors = [0.25, 0.5, 1.0]

for c in damping_factors:

    y = np.exp(-c*t)
    # Plot the function
    plt.plot(t, y)

plt.grid(True)
plt.legend([f'c={c}' for c in damping_factors])
plt.show()
```
